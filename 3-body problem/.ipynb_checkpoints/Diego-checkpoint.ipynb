```python
# We import our libs:
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Constants input:
a = float(input("Introduce a: "))
b = float(input("Introduce b: "))
N = float(input("Introduce N: "))
m1 = float(input("Introduce mass 1: "))
m2 = float(input("Introduce mass 2: "))
m3 = float(input("Introduce mass 3: "))
G = float(input("Introduce G: "))

# Vectors input:
r = []
v = []
temp = list(input("Introduce position vector 1: "))
r.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))
temp = list(input("Introduce position vector 2: "))
r.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))
temp = list(input("Introduce position vector 3: "))
r.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))
temp = list(input("Introduce velocity vector 1: "))
v.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))
temp = list(input("Introduce velocity vector 2: "))
v.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))
temp = list(input("Introduce velocity vector 3: "))
v.append(np.array([float(temp[i]) for i in range(len(temp)) if i%2 == 0]))

r = np.array(r)
v = np.array(v)

# Set constants:
h = (b - a) / N

# Print constants:
print("a: {}".format(a))
print("b: {}".format(b))
print("N: {}".format(N))
print("h: {}".format(h))

# Print vectors:
for i in range(3):
    print("Position vector r{}: {}".format(i+1, r[i]))
    print("Position vector v{}: {}".format(i+1, v[i]))
```

```python
def f(r1, r2, r3, m2, m3):
    return G * ((m2 * (r1 - r2) / np.linalg.norm((r1 - r2))**3) + (m3 * (r1 - r3) / np.linalg.norm((r1 - r3))**3))
```

```python
def Runge_Kutta(f, r, v):
    R1 = []
    R2 = []
    R3 = []
    V1 = []
    V2 = []
    V3 = []
    for i in range(int(N)):
        
        ### r1 vector:
        k1_r1_x = h * v[0, 0]
        k1_r1_y = h * v[0, 1]
        k1_r1_z = h * v[0, 2]

        l1_r1 = h * f(r[0], r[1], r[2], m2, m3)

        k2_r1_x = h * (v[0, 0] + 0.5*l1_r1[0])
        k2_r1_y = h * (v[0, 1] + 0.5*l1_r1[1])
        k2_r1_z = h * (v[0, 2] + 0.5*l1_r1[2])

        l2_r1 = h * f(r[0] + 0.5*k1_r1_x, r[1] + 0.5*k1_r1_y, r[2] + 0.5*k1_r1_z, m2, m3)

        k3_r1_x = h * (v[0, 0] + 0.5*l2_r1[0])
        k3_r1_y = h * (v[0, 1] + 0.5*l2_r1[1])
        k3_r1_z = h * (v[0, 2] + 0.5*l2_r1[2])

        l3_r1 = h * f(r[0] + 0.5*k2_r1_x, r[1] + 0.5*k2_r1_y, r[2] + 0.5*k2_r1_z, m2, m3)

        k4_r1_x = h * (v[0, 0] + l3_r1[0])
        k4_r1_y = h * (v[0, 1] + l3_r1[1])
        k4_r1_z = h * (v[0, 2] + l3_r1[2])

        l4_r1 = h * f(r[0] + k3_r1_x, r[1] + k3_r1_y, r[2] + k3_r1_z, m2, m3)


        ### r2 vector:
        k1_r2_x = h * v[1, 0]
        k1_r2_y = h * v[1, 1]
        k1_r2_z = h * v[1, 2]

        l1_r2 = h * f(r[1], r[0], r[2], m1, m3)

        k2_r2_x = h * (v[1, 0] + 0.5*l1_r2[0])
        k2_r2_y = h * (v[1, 1] + 0.5*l1_r2[1])
        k2_r2_z = h * (v[1, 2] + 0.5*l1_r2[2])

        l2_r2 = h * f(r[1] + 0.5*k1_r2_y, r[0] + 0.5*k1_r2_x, r[2] + 0.5*k1_r2_z, m1, m3)

        k3_r2_x = h * (v[1, 0] + 0.5*l2_r2[0])
        k3_r2_y = h * (v[1, 1] + 0.5*l2_r2[1])
        k3_r2_z = h * (v[1, 2] + 0.5*l2_r2[2])

        l3_r2 = h * f(r[1] + 0.5*k2_r2_y, r[0] + 0.5*k2_r2_x, r[2] + 0.5*k2_r2_z, m1, m3)

        k4_r2_x = h * (v[1, 0] + l3_r2[0])
        k4_r2_y = h * (v[1, 1] + l3_r2[1])
        k4_r2_z = h * (v[1, 2] + l3_r2[2])

        l4_r2 = h * f(r[1] + k3_r2_y, r[0] + k3_r2_x, r[2] + k3_r2_z, m1, m3)


        ### r3 vector:
        k1_r3_x = h * v[2, 0]
        k1_r3_y = h * v[2, 1]
        k1_r3_z = h * v[2, 2]

        l1_r3 = h * f(r[2], r[0], r[1], m1, m2)

        k2_r3_x = h * (v[2, 0] + 0.5*l1_r3[0])
        k2_r3_y = h * (v[2, 1] + 0.5*l1_r3[1])
        k2_r3_z = h * (v[2, 2] + 0.5*l1_r3[2])

        l2_r3 = h * f(r[2] + 0.5*k1_r3_z, r[0] + 0.5*k1_r3_x, r[1] + 0.5*k1_r3_y, m1, m2)

        k3_r3_x = h * (v[2, 0] + 0.5*l2_r3[0])
        k3_r3_y = h * (v[2, 1] + 0.5*l2_r3[1])
        k3_r3_z = h * (v[2, 2] + 0.5*l2_r3[2])

        l3_r3 = h * f(r[2] + 0.5*k2_r3_z, r[0] + 0.5*k2_r3_x, r[1] + 0.5*k2_r3_y, m1, m2)

        k4_r3_x = h * (v[2, 0] + l3_r3[0])
        k4_r3_y = h * (v[2, 1] + l3_r3[1])
        k4_r3_z = h * (v[2, 2] + l3_r3[2])

        l4_r3 = h * f(r[2] + k3_r3_z, r[0] + k3_r3_x, r[1] + k3_r3_y, m1, m2)

        # Runge-Kutta's step:
        
        # Position:
        x = r[0, 0] + (k1_r1_x + 2*k2_r1_x + 2*k3_r1_x + k4_r1_x)/6.
        y = r[0, 1] + (k1_r1_y + 2*k2_r1_y + 2*k3_r1_y + k4_r1_y)/6.
        z = r[0, 2] + (k1_r1_z + 2*k2_r1_z + 2*k3_r1_z + k4_r1_z)/6.
        if i > 0:
            R1.append(np.array([x, y, z]) + R1[i-1])
        else:
            R1.append(np.array([x, y, z]))
        x = r[1, 0] + (k1_r2_x + 2*k2_r2_x + 2*k3_r2_x + k4_r2_x)/6.
        y = r[1, 1] + (k1_r2_y + 2*k2_r2_y + 2*k3_r2_y + k4_r2_y)/6.
        z = r[1, 2] + (k1_r2_z + 2*k2_r2_z + 2*k3_r2_z + k4_r2_z)/6.
        if i > 0:
            R2.append(np.array([x, y, z]) + R2[i-1])
        else:
            R2.append(np.array([x, y, z]))
        x = r[2, 0] + (k1_r3_x + 2*k2_r3_x + 2*k3_r3_x + k4_r3_x)/6.
        y = r[2, 1] + (k1_r3_y + 2*k2_r3_y + 2*k3_r3_y + k4_r3_y)/6.
        z = r[2, 2] + (k1_r3_z + 2*k2_r3_z + 2*k3_r3_z + k4_r3_z)/6.
        if i > 0:
            R3.append(np.array([x, y, z]) + R3[i-1])
        else:
            R3.append(np.array([x, y, z]))
        
        # Velocity:
        vx = v[0, 0] + (l1_r1[0] + 2*l2_r1[0] + 2*l3_r1[0] + l4_r1[0])/6.
        vy = v[0, 1] + (l1_r1[1] + 2*l2_r1[1] + 2*l3_r1[1] + l4_r1[1])/6.
        vz = v[0, 2] + (l1_r1[2] + 2*l2_r1[2] + 2*l3_r1[2] + l4_r1[2])/6.
        if i > 0:
            V1.append(np.array([x, y, z]) + V1[i-1])
        else:
            V1.append(np.array([x, y, z]))
        vx = v[1, 0] + (l1_r2[0] + 2*l2_r2[0] + 2*l3_r2[0] + l4_r2[0])/6.
        vy = v[1, 1] + (l1_r2[1] + 2*l2_r2[1] + 2*l3_r2[1] + l4_r2[1])/6.
        vz = v[1, 2] + (l1_r2[2] + 2*l2_r2[2] + 2*l3_r2[2] + l4_r2[2])/6.
        if i > 0:
            V2.append(np.array([x, y, z]) + V2[i-1])
        else:
            V2.append(np.array([x, y, z]))
        vx = v[2, 0] + (l1_r3[0] + 2*l2_r3[0] + 2*l3_r3[0] + l4_r3[0])/6.
        vy = v[2, 1] + (l1_r3[1] + 2*l2_r3[1] + 2*l3_r3[1] + l4_r3[1])/6.
        vz = v[2, 2] + (l1_r3[2] + 2*l2_r3[2] + 2*l3_r3[2] + l4_r3[2])/6.
        if i > 0:
            V3.append(np.array([x, y, z]) + V3[i-1])
        else:
            V3.append(np.array([x, y, z]))
    
    R = np.array([np.array(R1), np.array(R2), np.array(R3)])
    V = np.array([np.array(V1), np.array(V2), np.array(V3)])
    return R, V
```

```python
R, V = Runge_Kutta(f, r, v)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(R[0,0], R[0,1], R[0,2], label='Object 1')
ax.plot(R[1,0], R[1,1], R[1,2], label='Object 2')
ax.plot(R[2,0], R[2,1], R[2,2], label='Object 3')
ax.legend()
plt.show()
```
